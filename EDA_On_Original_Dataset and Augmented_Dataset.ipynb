```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python

t_healthy = '/content/drive/MyDrive/Dataset-3/Test/Healthy'
t_scab='/content/drive/MyDrive/Dataset-3/Test/Scab'
tr_healthy='/content/drive/MyDrive/Dataset-3/Train/Healthy'
tr_scab='/content/drive/MyDrive/Dataset-3/Train/Scab'
num_healthy_images = len(os.listdir(t_healthy))+len(os.listdir(tr_healthy))
num_scab_images = len(os.listdir(t_scab))+len(os.listdir(tr_scab))
plt.figure(figsize=(10, 6))
data = {'Healthy': num_healthy_images, 'Scab':num_scab_images}
plt.bar(data.keys(), data.values(), color = ['green', 'red'])
plt.title('Bar Plot for Classes (Original Dataset)', fontsize=14)
plt.xlabel("Classes", fontsize=17) 
plt.ylabel("Count", fontsize=16) 

plt.savefig('data_bar_plot', facecolor= 'w', transparent= False, bbox_inches= 'tight', dpi= 300)
print('Original Dataset\n')
print('Healthy:', num_healthy_images)
print('Scab:', num_scab_images)
```

```python
# Training set

plt.figure(figsize=(10, 6))
data = {'Healthy': len(os.listdir(tr_healthy)), 'Scab': len(os.listdir(tr_scab))}
plt.bar(data.keys(), data.values(), color = ['green', 'yellow'])
plt.title('Bar Plot for Train Set', fontsize=14)
plt.xlabel("Classes", fontsize=12) 
plt.ylabel("Count", fontsize=12) 

plt.savefig('train_set_plot', facecolor= 'w', transparent= False, bbox_inches= 'tight', dpi= 300)
```

```python
# Test set

plt.figure(figsize=(10, 6))
data = {'Healthy': len(os.listdir(t_healthy)), 'Scab': len(os.listdir(t_scab))}
plt.bar(data.keys(), data.values(), color = ['red', 'green'])
plt.title('Bar Plot for Test Set', fontsize=14)
plt.xlabel("Classes", fontsize=12) 
plt.ylabel("Count", fontsize=12) 

plt.savefig('test_set_plot', facecolor= 'w', transparent= False, bbox_inches= 'tight', dpi= 300)
```
